```python
# 多图布局
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建一个2x2的子图布局
plt.figure(figsize=(10, 6))  # 设置整个图的大小

# 第一个子图
plt.subplot(2, 2, 1)  # 第一个子图位于2x2布局的第1个位置
plt.plot(x, y1, 'r-')  # 绘制sin曲线
plt.title('Sin Function')  # 设置标题

# 第二个子图
plt.subplot(2, 2, 2)  # 第二个子图位于2x2布局的第2个位置
plt.plot(x, y2, 'g-')  # 绘制cos曲线
plt.title('Cos Function')  # 设置标题

# 第三个子图
plt.subplot(2, 2, 3)  # 第三个子图位于2x2布局的第3个位置
plt.plot(x, y1 + y2, 'b-')  # 绘制sin+cos曲线
plt.title('Sin + Cos Function')  # 设置标题

# 第四个子图
plt.subplot(2, 2, 4)  # 第四个子图位于2x2布局的第4个位置
plt.plot(x, y1 - y2, 'm-')  # 绘制sin-cos曲线
plt.title('Sin - Cos Function')  # 设置标题

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt # 图形嵌套
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100)
y = np.sin(x)
# 创建外层图形
fig = plt.figure(figsize=(8, 6))
# 在外层图形中创建主要的子图
plt.plot(x, y, 'y-')
plt.title('Main Plot')
# 在外层图形中创建嵌套的子图
left, bottom, width, height = [0.4, 0.4, 0.25, 0.25]
axes = fig.add_axes([left, bottom, width, height]) #创建一个新的坐标轴（axes）
axes.plot(x, y, 'r-', label='Sin Function (Inset)')
plt.title('Inset Plot')
# axes.set_title('Inset Plot')
# 显示图例
plt.legend()
# 显示图形
plt.show()
```

```python
# 生成数据便于后续绘图
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 创建数据
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
students = ['张三']
subjects = ['语文', '数学', '英语']
# 生成随机成绩数据
data = np.random.randint(60, 100, size=(12, len(subjects)))  # 12个月，3门科目
# 将数据转换为 DataFrame
df = pd.DataFrame(data, columns=subjects, index=months)
# 将数据写入 Excel 表格
excel_file = '张三成绩.xlsx'
df.to_excel(excel_file)
print("数据已保存到 Excel 文件:", excel_file)
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))

x = df.index # 获取月份作为 x 轴数据
y1 = df['语文'] # 获取张三在语文科目上的成绩作为 y 轴数据
y2 = df['数学']
y3 = df['英语']

plt.plot(x, y1, marker='o', color="r",ls='--',label="语文")
plt.plot(x, y2, marker='*', color="b",ls='-',label="数学")
plt.plot(x, y3, marker='v', color="y",ls='-',label="英语")

plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel('分数')
plt.legend() # 添加图例
plt.grid(True) # 网格线

# 显示图形
plt.xticks(rotation=45) # 将 X 轴刻度标签的角度旋转为 45 度
plt.tight_layout()  # 自动调整子图参数，以使所有子图适应图形区域
plt.show()
```

```python
# 绘制多个折线图
plt.figure(figsize=(10, 6))

for subject in subjects:
    # print(subject)
    plt.plot(df.index, df[subject], marker='o', label=subject) # 循环绘制，df.index为月份、subject为语文 数学 英语

plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel('分数')
plt.legend()
plt.grid(True)

# 显示图形
plt.xticks(rotation=45) # 将 X 轴刻度标签的角度旋转为 45 度
plt.tight_layout()  # 自动调整子图参数，以使所有子图适应图形区域
plt.show()
```

```python
# 绘制多个柱状图
import pandas as pd
import matplotlib.pyplot as plt
# 读取 Excel 表格数据
excel_file = '张三成绩.xlsx'
df = pd.read_excel(excel_file, index_col=0)
# 绘制柱状图
plt.figure(figsize=(12, 6))
# 获取月份
months = df.index
# 设置每个月份的位置
x = range(len(months)) # x=range(12)
# 设置柱状图宽度
bar_width = 0.25
# 绘制张三在三门科目上的成绩柱状图 x=[0,1,2,3...]  \\\ i + bar_width for i in x = 0.25,1.25,2.25 \\\i + 2 * bar_width for i in x = 0.5,1.5,2.5...
plt.bar(x, df['语文'], width=bar_width, align='center', label='语文', color='skyblue') # align='center'  柱子对齐方式为居中
plt.bar([i + bar_width for i in x], df['数学'], width=bar_width, align='center', label='数学', color='r') #x 轴的位置，即月份，通过偏移量实现柱子的横向偏移
plt.bar([i + 2 * bar_width for i in x], df['英语'], width=bar_width, align='center', label='英语', color='lightgreen')
# 设置图形标题和坐标轴标签
plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel(' ')
# 设置 X 轴刻度标签
plt.xticks(x, months, rotation=45)  # 必须设置
# 添加图例
plt.legend()
# 显示图形
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt # 绘制散点图
import numpy as np
# 生成样本数据
np.random.seed(21)
height = np.random.normal(loc=170, scale=10, size=100)  # 平均身高为170cm，标准差为10cm
# height = np.random.randn(100) # 默认生成为均值为0.标准差为1
weight = 0.6 * height + np.random.normal(loc=0, scale=5, size=100)  # 体重与身高的线性关系，斜率为0.6，截距为0，标准差为5kg
# 绘制散点图
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='skyblue', alpha=0.9)  #  alpha 散点的透明度
# 添加标题和坐标轴标签 
plt.title('身高与体重关系散点图')
plt.xlabel('身高 (cm)')
plt.ylabel('体重 (kg)')
# 显示图形
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# 绘制饼图
# 读取 Excel 表格数据
excel_file = '湖南10天天气情况.xlsx'
df = pd.read_excel(excel_file)
# 绘制饼图
plt.figure(figsize=(6, 7))
# labels每个扇形对应的标签，即日期 ， autopct每个扇形中显示的百分比格式， startangle 绘制的起始位置
plt.pie(df['Temperature'], labels=df['Date'], autopct='%.2f%%', startangle=90) # %f %d 格式化输出的
# 添加标题
plt.title('湖南10天天气情况饼图')
# 显示图形
plt.axis('equal')  # 保证饼图是圆的
plt.show()
```

```python
import matplotlib.pyplot as plt
# 创建一个图形和一个轴对象
fig, ax = plt.subplots()
# 定义水果种类和对应的供应量
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
# 每个柱子的标签和颜色
bar_labels = ['red', 'blue', '_red', 'orange']  # 标签用于图例
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']  # 颜色用于柱状图
# 绘制柱状图
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
# 设置 y 轴标签和图形标题
ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
# 添加图例，并设置标题
ax.legend(title='Fruit color')
# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

ax = plt.figure().add_subplot(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8,
                alpha=0.3)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='y', offset=40, cmap='coolwarm')

ax.set(xlim=(-40, 40), ylim=(-40, 40), zlim=(-100, 100),
       xlabel='X', ylabel='Y', zlabel='Z')

plt.show()
```
