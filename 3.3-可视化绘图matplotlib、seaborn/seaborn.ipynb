```python
# 
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
# 
f = flights.pivot(index="month", columns="year", values="passengers")
# 
plt.figure(figsize=(10, 8))  # 
sns.heatmap(f, annot=True, fmt="d",cmap="YlGnBu")  # annot=True fmt="d" 
plt.title("Flights Passenger Count by Year and Month")  # 
plt.xlabel("Year")  #  x 
plt.ylabel("Month")  #  y 
plt.show()  # 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f = flights.pivot(index="month", columns="year", values="passengers")
# 
plt.figure(figsize=(10, 10))  # 
sns.heatmap(f.corr(), annot=True,cmap='coolwarm')
```
