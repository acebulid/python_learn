```python
# 热力图
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
# 将长格式数据转换为宽格式数据，以便绘制热力图
f = flights.pivot(index="month", columns="year", values="passengers")
# 绘制热力图
plt.figure(figsize=(10, 8))  # 设置图表大小
sns.heatmap(f, annot=True, fmt="d",cmap="YlGnBu")  # annot=True 在单元格中显示数据，fmt="d" 格式化为整数
plt.title("Flights Passenger Count by Year and Month")  # 设置标题
plt.xlabel("Year")  # 设置 x 轴标签
plt.ylabel("Month")  # 设置 y 轴标签
plt.show()  # 显示图表
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f = flights.pivot(index="month", columns="year", values="passengers")
# 生成相关系数矩阵的热力图
plt.figure(figsize=(10, 10))  # 设置图表大小
sns.heatmap(f.corr(), annot=True,cmap='coolwarm')
```
